# Módulos

Los modulos son la unidad de programa mas grande en Python, y normalemnte son archivos python externos, aunque puede también hacer referencia a archivos de otros lenguajes como C, C#, Java, entre otros.

> Un ejemplo de módulos que usan otros praquetes es Pandas, que usa Numpy que asu vez usa programas escritos en C++ y Fortran

la forma de declarar modulos es

```python
import
Lets a client (importer) fetch a module as a whole
from
Allows clients to fetch particular names from a module
```

en donde ``import`` trae todo lo correspondiente de un modulo y cuando lo complementamos con ``as`` lo que decidimos hacer es cambiar el nombre con el que haremos referencias, adicional si no queremos todo el módulo, podemos especificar exactamente que funciones, clases o variables queremos traer con el ``from``

## Por qué usar módulos

Las principales razones por las que usar módulos son:

- Reutilización de código (uno de los zen de Python)
- Organizar los nombres y asignaciones en nuestros programas
- Implementar servicios o datos que pueden ser compartidos con otros módulos

## Estructuración de un programa en Python

Un programa de Python normalmente se compone de un archivo main o principal que es el usado para inciar la aplicación, sea cual sea la aplicación, este archivo a su vez invocara otros modulos muchas veces personalizados y propiedad del desarrollador de la aplicación y estos usarán aun mas módulos de terceros o includios en Python.

## Como funciona la importación en Python

En Python las importanciones no es como en un lenguaje compilado, en donde se inserta el código compilado en el archivo final, sino que realmente ejecuta una pieza de código encargada de gestionar esa nueva referencia en una nueva variable, este tiene los siguientes pasos:
1. encuentra el archivo del modulo 
2. lo compile a Bytecode si es necesario
3. corre el código del modulo para crear el objeto relacionado

## Donde busca los madulos Python

1. The home directory of the program
2. PYTHONPATH directories (if set)
3. Standard library directories
4. The contents of any .pth files (if present)
5. The site-packages home of third-party extensions

# Module Coding Basics

Los módulos cumplen con las siguientes caractreristicas

- se cargan como variables, por lo que siguen las mismas reglas de nombres de variables en Python
- dentro de los módulos se cada uno tiene un ambito de alcance para variables y funciones
- el ``import`` y ``from`` hacen el mismo proceso, solo que el ``from`` copia las referencias del módulo antes de eliminar las referencias a todo el módulo

> el ``from`` tiene la siguente analogía

```python
from module import x,y
x = module.x
y = module.y
```
# Modules Package

Hasta ahora hemos visto solo que los módulos son un archivo de Python o incluso de otros lenguajes de prgramación, pero es posible crear carpetas de modulos, también conocidos como paquetes, dichos paquetes es un agrupamiento de módulos (archivos) que tienen algo o un fin en común.

para importar los módulos de un paque seguimos la estructura de carpetas que manejan

```python
import dir1.dir2.module
from dir1.dir2.module import x,y
```

además de tener en cuenta el procesamiento de los datos, el siguiente paso será crear un archivo de inicialización dentro de cada directorio, este archivo se debe de llamar ``__init__.py`` y la principal caracteristica de este archivo es que funciona como inicializador del paquete, es decir que nos permite ejecutar tareas necesarias una vez de llama al módulo, como por ejemplo conectarse con una base de datos.

Cuando usamos pauetes, es posible realizar importanciones de módulos con paths relativos o absolutos, del mismo modo que se hace con las rutas de directores en el OS

```python
from .dir1.module import x,y == from dir0.dir1.module import x,y
```
